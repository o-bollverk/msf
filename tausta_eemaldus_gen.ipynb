```python
import os
from os import path

def bash_generator(filename, loc1):
    file1 = open(filename,"a") 
    file1.write("#!/bin/sh \n")
    for i in os.listdir(loc1):
        for j in os.listdir(loc1 + "/" + i):
                output_name = loc1 + "/" + i + "/" +  j 
                output_name2 =  output_name[:len(output_name)-4] + ".png"
                if path.exists(output_name2) == False and j[len(j)-4:] == ".jpg": 
                    file1.write("rembg -o" + " '"  + output_name2 + "' " + "'"   + loc1 + "/" + i + "/" + j + "'" +  "\n")
    file1.close() 
    
loc1 = "/home/revilo/Desktop/msf_local/msf/random_selection_copy/random_selection_train"
bash_generator(filename = "tausta_eemaldus_train.txt", loc1 = loc1)

loc2 = "/home/revilo/Desktop/msf_local/msf/random_selection_copy/random_selection_test"
bash_generator(filename= "tausta_eemaldus_test.txt", loc1 = loc2)
```
